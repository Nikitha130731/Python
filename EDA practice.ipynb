```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('Data_Train.xlsx')
```

```python
train_df.head()
```

```python
test_df=pd.read_excel('Test_set.xlsx')
```

```python
test_df.head()
```

```python
#concatenation using pandas
final_df= pd.concat([train_df,test_df])
```

```python
final_df.head()
```

```python
final_df.describe()
```

```python
final_df.info()
```

```python
final_df.isnull().sum()
```

```python
# feature engineering
# splitting data of date
final_df['Date_of_Journey'].str.split('/')
```

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
```

```python
final_df['Date']
```

```python
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
```

```python
final_df['Month']
```

```python
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df['Year']
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
# converting object to int data
final_df['Date']= final_df['Date'].astype(int)
final_df['Month']= final_df['Month'].astype(int)
final_df['Year']= final_df['Year'].astype(int)
```

```python
final_df.tail().info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Time'].str.split(' ')
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
final_df['Arrival_Time']
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].apply(lambda x:x.split(':')[0]).astype(int)
final_df['Arrival_minutes']=final_df['Arrival_Time'].apply(lambda x:x.split(':')[1]).astype(int)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Dep_Time']=final_df['Dep_Time'].str.split(':')
```

```python
final_df['Dep_Time']
```

```python
final_df['Dep_hour']=final_df['Dep_Time'].str[0].astype(int)
final_df['Dep_minutes']=final_df['Dep_Time'].apply(lambda x:x[1]).astype(int)
```

```python
final_df.head(1)
```

```python
final_df.info()
```

```python
final_df.drop('Dep_Time', axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df.head(1)
```

```python


# One-liner with enhanced error handling
final_df['Total_Duration_Minutes'] = final_df['Duration'].fillna('0h, 0m').astype(str).apply(
    lambda x: sum(int(i[:-1]) * 60 if 'h' in i else int(i[:-1]) for i in x.replace(' ', '').replace("'", "").split(',') if i and (len(i) > 1 and (i[-1] == 'm' or i[-1] == 'h')))
)


```

```python
final_df['Duration']=final_df['Duration'].fillna('0h, 0m').astype(str)
```
